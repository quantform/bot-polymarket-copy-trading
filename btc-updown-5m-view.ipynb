```typescript
import { display } from 'tslab';
import { marked } from 'marked';
import { writeFileSync } from 'fs';
import { execSync } from 'child_process';
import { Module, core, logger } from '@quantform/core';

function svg(spec: unknown) {
  writeFileSync('./.jupiter/spec.vl.json', JSON.stringify(spec));

  const svg = execSync('npx vl2svg ./.jupiter/spec.vl.json', {
    encoding: 'utf8'
  });

  return display.html(svg);
}

async function md(text: string) {
  return display.html(await marked(text));
}

const module = new Module([
  ...core(),
  logger({
    for: context => ({
      info: () => {},
      debug: () => {},
      warn: () => {},
      error: () => console.log(context)
    })
  })
]);

const { act } = await module.awake();
```

```typescript
import { lastValueFrom, switchMap, map, tap } from 'rxjs';
import { queryMarket } from './src/polymarket/query-market';
import { queryUserActivity } from './src/polymarket/query-user-activity';

const userWallet = '0x1d0034134e339a309700Ff2D34e99FA2d48b0313';
const timestamp = 1772020800;

//0x1d0034134e

const userActivity: any[] = await act(() =>
  lastValueFrom(
    queryMarket(`btc-updown-5m-${timestamp}`).pipe(
      switchMap(it => queryUserActivity(it.conditionId, userWallet)),
      map((it: any[]) => it.filter(it => it.type == 'TRADE'))
    )
  )
);
```

```typescript
function buildExecutionSpec(userActivity) {
  const start = timestamp;

  const seconds = 300;

  const payoutUp = new Array(seconds).fill(0);
  const payoutDown = new Array(seconds).fill(0);
  const costUp = new Array(seconds).fill(0);
  const costDown = new Array(seconds).fill(0);

  for (const a of userActivity) {
    const sec = Math.floor(a.timestamp - start);
    if (sec < 0 || sec >= seconds) continue;

    if (a.outcome === 'Up') {
      payoutUp[sec] += a.size;
      costUp[sec] += a.size * a.price;
    } else {
      payoutDown[sec] += a.size;
      costDown[sec] += a.size * a.price;
    }
  }

  // cumulative
  for (let i = 1; i < seconds; i++) {
    payoutUp[i] += payoutUp[i - 1];
    payoutDown[i] += payoutDown[i - 1];
    costUp[i] += costUp[i - 1];
    costDown[i] += costDown[i - 1];
  }

  const values = [];

  for (let i = 0; i < seconds; i++) {
    const totalCost = costUp[i] + costDown[i];
    const riskUp = payoutUp[i] - totalCost;
    const riskDown = payoutDown[i] - totalCost;

    values.push({
      sec: i,
      payoutUp: payoutUp[i],
      payoutDown: payoutDown[i],
      totalCost,
      riskUp,
      riskDown
    });
  }

  return {
    $schema: 'https://vega.github.io/schema/vega-lite/v5.json',
    vconcat: [
      {
        width: 1000,
        height: 300,
        title: 'Payout vs Cost',
        data: { values },
        layer: [
          {
            mark: 'line',
            encoding: {
              x: { field: 'sec', type: 'quantitative', title: 'Seconds' },
              y: { field: 'payoutUp', type: 'quantitative', title: 'USD' },
              color: { value: '#1f77b4' }
            }
          },
          {
            mark: 'line',
            encoding: {
              x: { field: 'sec', type: 'quantitative' },
              y: { field: 'payoutDown', type: 'quantitative' },
              color: { value: '#d62728' }
            }
          },
          {
            mark: { type: 'line', strokeDash: [4, 4] },
            encoding: {
              x: { field: 'sec', type: 'quantitative' },
              y: { field: 'totalCost', type: 'quantitative' },
              color: { value: '#000000' }
            }
          }
        ]
      },
      {
        width: 1000,
        height: 300,
        title: 'Risk Surface (PnL if resolves now)',
        data: { values },
        layer: [
          {
            mark: 'line',
            encoding: {
              x: { field: 'sec', type: 'quantitative' },
              y: { field: 'riskUp', type: 'quantitative', title: 'PnL' },
              color: { value: '#1f77b4' }
            }
          },
          {
            mark: 'line',
            encoding: {
              x: { field: 'sec', type: 'quantitative' },
              y: { field: 'riskDown', type: 'quantitative' },
              color: { value: '#d62728' }
            }
          },
          {
            mark: { type: 'rule', color: 'gray' },
            encoding: {
              y: { datum: 0 }
            }
          }
        ]
      }
    ]
  };
}

svg(buildExecutionSpec(userActivity));
```

```typescript
function buildStageSpec(userActivity) {
  const start = timestamp;
  const seconds = 300;

  const payoutUp = new Array(seconds).fill(0);
  const payoutDown = new Array(seconds).fill(0);

  for (const a of userActivity) {
    const sec = Math.floor(a.timestamp - start);
    if (sec < 0 || sec >= seconds) continue;

    if (a.outcome === 'Up') {
      payoutUp[sec] += a.size;
    } else {
      payoutDown[sec] += a.size;
    }
  }

  for (let i = 1; i < seconds; i++) {
    payoutUp[i] += payoutUp[i - 1];
    payoutDown[i] += payoutDown[i - 1];
  }

  const values = [];

  for (let i = 0; i < seconds; i++) {
    const up = payoutUp[i];
    const down = payoutDown[i];

    const skew = up - down;

    const hedge = Math.max(up, down) === 0 ? 0 : Math.min(up, down) / Math.max(up, down);

    values.push({
      sec: i,
      skew,
      hedge
    });
  }

  return {
    $schema: 'https://vega.github.io/schema/vega-lite/v5.json',
    vconcat: [
      {
        width: 1000,
        height: 300,
        title: 'Directional Exposure (Inventory Skew)',
        data: { values },
        mark: 'line',
        encoding: {
          x: { field: 'sec', type: 'quantitative' },
          y: { field: 'skew', type: 'quantitative', title: 'Up - Down' }
        }
      },
      {
        width: 1000,
        height: 300,
        title: 'Hedge Ratio (Stage Detector)',
        data: { values },
        layer: [
          {
            mark: 'line',
            encoding: {
              x: { field: 'sec', type: 'quantitative' },
              y: { field: 'hedge', type: 'quantitative', title: 'Hedge Ratio' }
            }
          },
          {
            mark: { type: 'rule', strokeDash: [4, 4] },
            encoding: { y: { datum: 0.3 } }
          },
          {
            mark: { type: 'rule', strokeDash: [4, 4] },
            encoding: { y: { datum: 0.8 } }
          }
        ]
      }
    ]
  };
}

svg(buildStageSpec(userActivity));
```
